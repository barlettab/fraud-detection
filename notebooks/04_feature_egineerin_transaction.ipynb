Como seria um comportamento legítimo?

Exemplos no seu dataset:

um cartão compra com valores “parecidos” ao longo do tempo
um usuário usa sempre o mesmo email
compras têm certa regularidade
valores não variam absurdamente

> Isso define o “baseline normal”

Fraude quase sempre é:

comportamento fora do padrão
mudança brusca
repetição excessiva
inconsistência entre variáveis

Exemplos:

valor MUITO maior que o normal do cartão
muitos pedidos em pouco tempo
email diferente do usual
cartão aparecendo em excesso

> aqui nasce a ideia de “sinal de fraude”

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/train_transaction.csv")
```

```python
# TransactionDT
df["hour"] = (df["TransactionDT"] // 3600) % 24
df["day"] = df["TransactionDT"] // 86400
df["week"] = df["TransactionDT"] // (86400 * 7)
```

```python
# TransactionAmt
df["log_amount"] = np.log1p(df["TransactionAmt"])
```

```python
# Card1
df["card1_freq"] = df["card1"].map(df["card1"].value_counts())
```

```python
# Pergunta: Esse cartão sempre aparece nessa região?
# Ou seja, qual a frequência de cada combinação de card1 e addr1?
df["card1_addr_freq"] = df.groupby(["card1", "addr1"]).transform("size")
```

```python
# Pergunta: Esse cartão normalmente transaciona perto ou longe?
df["card1_mean_dist"] = df.groupby("card1")["dist1"].transform(lambda x: x.mean())
```

```python
# Pergunta: A transação está fora do padrão geográfico do cartão?
df["dist1_vs_card1_mean"] = df["dist1"] - df.groupby("card1")["dist1"].transform("mean")
```
