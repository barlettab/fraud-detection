```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 200)
```

```python
df = pd.read_csv("../data/train_transaction.csv")
print(df.shape)
df.head()
```

```python
df.info()
df.isnull().mean().sort_values(ascending=False).head(20)
```

```python
# Comportamento Temporal
# Hipótese: A fraude muitas vezes acontece em horários atípicos, como de madrugada.
# Pergunta: A transação aconteceu rápido demais?
temporal_features = {}

temporal_features["hour"] = (df["TransactionDT"] // 3600) % 24
temporal_features["day"] = df["TransactionDT"] // 86400
temporal_features["week"] = df["TransactionDT"] // (86400 * 7)
```

```python
# Features de velocidade em que ocorre as transações
df = df.sort_values(["card1", "TransactionDT"])

temporal_features["time_since_last"] = (
    df.groupby("card1")["TransactionDT"].diff()
)

temporal_features["fast_txn_flag"] = (
    temporal_features["time_since_last"] < 300
).astype(int)

df = pd.concat(
    [df, pd.DataFrame(temporal_features, index=df.index)],
    axis=1
)
```

```python
df["dist1"] = df["dist1"].fillna(-1)

# Comportamento Geográfico
# Hipótese: Transações fraudulentas frequentemente apresentam inconsistências de localização e distância geográfica.
# Pergunta: A transação aconteceu em um local incomum para o usuário?
geo_features = {}

geo_features["card1_addr_freq"] = (
    df.groupby(["card1", "addr1"])["card1"]
    .transform("count")
)

geo_features["card1_mean_dist"] = (
    df.groupby("card1")["dist1"]
    .transform("mean")
)

geo_features["dist1_vs_card1_mean"] = (
    df["dist1"] - geo_features["card1_mean_dist"]
)

previous_addr = df.groupby("card1")["addr1"].shift(1)

geo_features["addr_change"] = (
    (df["addr1"] != previous_addr) &
    previous_addr.notnull()
).astype(int)

geo_features["dist1_missing"] = df["dist1"].isnull().astype(int)

df = pd.concat(
    [df, pd.DataFrame(geo_features, index=df.index)],
    axis=1
)
```

```python
# Comportamento Monetário
# Hipótese: Transações fraudulentas podem apresentar padrões de valor atípicos.
# Pergunta: O valor da transação é incomum para o usuário?
amount_features = {}

amount_features["log_amount"] = np.log1p(df["TransactionAmt"])

amount_features["card1_amt_mean"] = (
    df.groupby("card1")["TransactionAmt"]
    .transform("mean")
)

amount_features["card1_amt_std"] = (
    df.groupby("card1")["TransactionAmt"]
    .transform("std")
)

amount_features["amt_vs_card1_mean"] = (
    df["TransactionAmt"] - amount_features["card1_amt_mean"]
)

amount_features["amt_zscore_card1"] = (
    amount_features["amt_vs_card1_mean"] /
    (amount_features["card1_amt_std"] + 1e-6)
)

df = pd.concat(
    [df, pd.DataFrame(amount_features, index=df.index)],
    axis=1
)
```

```python
# Comportamento Identificador
# Hipótese: Fraudes muitas vezes envolvem inconsistências nos atributos de identificação.
# Pergunta: A identidade parece inconsistente?
identity_features = {}

identity_features["card1_freq"] = (
    df["card1"].map(df["card1"].value_counts())
)

identity_features["email_domain_change"] = (
    df["P_emaildomain"] != df["R_emaildomain"]
).astype(int)

df = pd.concat(
    [df, pd.DataFrame(identity_features, index=df.index)],
    axis=1
)
```

```python
df["card4"] = df["card4"].astype("category")
df["card6"] = df["card6"].astype("category")
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df["amt_zscore_card1"] = (
    df["amt_zscore_card1"]
    .replace([np.inf, -np.inf], np.nan)
    .fillna(0)
)
```

A análise dos valores ausentes das novas features mostrou que o pipeline de feature engineering apresentou boa consistência geral, já que a maioria das variáveis criadas não possui valores nulos. As features temporais (hour, day, week e fast_txn_flag) e monetárias (log_amount e amt_vs_card1_mean) foram geradas integralmente, indicando que as transformações aplicadas sobre TransactionDT e TransactionAmt foram bem-sucedidas. 

Já a feature time_since_last apresentou cerca de 2,3% de valores ausentes, o que é esperado, pois a primeira transação de cada cartão não possui uma transação anterior para comparação temporal. A variável amt_zscore_card1 apresentou apenas 0,58% de valores nulos, provavelmente associados a cartões com poucas transações, dificultando o cálculo do desvio padrão. A feature card1_addr_freq foi a que apresentou maior percentual de valores ausentes (11,1%), possivelmente devido à ausência de informações em addr1 para algumas transações.

De forma geral, os resultados indicam que as novas features foram construídas com boa qualidade e baixo nível de perda de informação, tornando o conjunto de dados mais rico para a etapa de modelagem preditiva de fraude.

#### Behavioral Consistency Features
Fraudulent behavior often combines abnormal timing, monetary deviation and geographic inconsistency.

```python
consistency_features = {}

# Pergunta:
# O valor atual é muito diferente da transação anterior?

previous_amount = (
    df.groupby("card1")["TransactionAmt"]
    .shift(1)
)

consistency_features["amt_ratio_prev"] = (
    df["TransactionAmt"] /
    (previous_amount + 1e-6)
)

consistency_features["amt_ratio_prev"] = (
    consistency_features["amt_ratio_prev"]
    .replace([np.inf, -np.inf], np.nan)
    .fillna(1)
)

# Pergunta:
# O valor atual é muito diferente da média das últimas 5 transações?

rolling_mean_5 = (
    df.groupby("card1")["TransactionAmt"]
    .transform(
        lambda x: x.shift(1).rolling(5).mean()
    )
)

consistency_features["amt_ratio_last5"] = (
    df["TransactionAmt"] /
    (rolling_mean_5 + 1e-6)
)

consistency_features["amt_ratio_last5"] = (
    consistency_features["amt_ratio_last5"]
    .replace([np.inf, -np.inf], np.nan)
    .fillna(1)
)

# Pergunta:
# Esse cartão transacionou rápido demais?

consistency_features["high_velocity_flag"] = (
    df["time_since_last"] < 300
).astype(int)

# Pergunta:
# O cartão mudou de endereço muito rapidamente?

consistency_features["rapid_addr_change"] = (
    (df["time_since_last"] < 300) &
    (df["addr_change"] == 1)
).astype(int)

# Pergunta:
# A compra foi muito rápida e o valor é estranho?

consistency_features["rapid_large_txn"] = (
    (df["time_since_last"] < 300) &
    (df["amt_zscore_card1"] > 3)
).astype(int)

# Pergunta:
# Houve mudança geográfica e valor estranho?

consistency_features["geo_amount_anomaly"] = (
    (df["addr_change"] == 1) &
    (df["amt_zscore_card1"] > 3)
).astype(int)

df = pd.concat(
    [df, pd.DataFrame(consistency_features, index=df.index)],
    axis=1
)
```
