# 1. Business Understanding

## 1.1 Introdução

A fraude financeira é um dos principais desafios enfrentados por bancos, fintechs e plataformas de pagamento digital. Com o crescimento das transações online e da digitalização dos serviços financeiros, aumentou também a sofisticação dos mecanismos utilizados para fraudar sistemas de pagamento, roubo de identidade e validação de transações.

Nesse contexto, sistemas de detecção de fraude baseados em Machine Learning tornaram-se fundamentais para identificar comportamentos suspeitos em tempo real, reduzindo perdas financeiras e protegendo usuários legítimos.

Este projeto utiliza o dataset da IEEE-CIS Fraud Detection Competition, uma competição desenvolvida em parceria entre a IEEE Computational Intelligence Society (IEEE-CIS) e a empresa Vesta Corporation, especializada em soluções antifraude.

O objetivo principal é desenvolver modelos capazes de prever a probabilidade de uma transação ser fraudulenta com base em informações transacionais, comportamentais e de identidade.


## 1.2 O Problema de Negócio

Instituições financeiras processam milhões de transações diariamente. Dentro desse volume, uma pequena parcela representa atividades fraudulentas, como:

* uso indevido de cartões
* roubo de identidade
* transações não autorizadas
* manipulação de contas
* compras realizadas por agentes maliciosos

O principal desafio é que fraudes representam apenas uma pequena fração do total de operações, tornando o problema altamente desbalanceado e difícil de detectar.

Além disso, o sistema antifraude precisa equilibrar segurança e experiência do usuário. Um sistema excessivamente rígido pode bloquear clientes legítimos, enquanto um sistema permissivo pode permitir que fraudes passem despercebidas.

Dessa forma, a detecção de fraude não é apenas um problema técnico, mas também um problema estratégico de gerenciamento de risco.


## 1.3 Tipos de Erro em Detecção de Fraude

Em problemas de classificação de fraude, existem dois erros principais:

### 1.3.1 Falso Positivo (False Positive)

Ocorre quando uma transação legítima é classificada como fraude.

Impactos:

* bloqueio indevido de clientes
* perda de confiança
* aumento de atrito na experiência do usuário
* custos operacionais de revisão manual


### 1.3.2 Falso Negativo (False Negative)

Ocorre quando uma transação fraudulenta é classificada como legítima.

Impactos:

* perdas financeiras diretas
* aumento de chargebacks
* prejuízo reputacional
* vulnerabilidade do sistema antifraude


## 1.4 Trade-off de Risco

Sistemas antifraude operam constantemente em um equilíbrio entre segurança e usabilidade.

**Um modelo extremamente conservador pode reduzir fraudes, mas também aumentar significativamente os falsos positivos.** Por outro lado, um modelo muito permissivo melhora a experiência do usuário, porém aumenta o risco financeiro da operação.

Por esse motivo, a escolha de métricas e limiares de decisão é fundamental no desenvolvimento de modelos de fraude.

Neste projeto, a análise será conduzida considerando que:

* fraudes possuem alto custo financeiro
* o dataset é altamente desbalanceado
* diferentes tipos de erro possuem impactos distintos no negócio


## 1.5 Entendimento do Dataset

O projeto utiliza quatro arquivos principais:

| Dataset                 | Descrição                                              |
| ----------------------- | ------------------------------------------------------ |
| `train_transaction.csv` | Dados transacionais utilizados para treinamento        |
| `train_identity.csv`    | Informações complementares de identidade e dispositivo |
| `test_transaction.csv`  | Dados transacionais utilizados para inferência         |
| `test_identity.csv`     | Informações de identidade do conjunto de teste         |

Os datasets de transação e identidade são relacionados através da variável `TransactionID`.

**Nem todas as transações possuem informações de identidade associadas, refletindo um cenário realista em sistemas financeiros, onde parte das informações do usuário pode estar indisponível.**

## 1.6 Estrutura das Features

O dataset contém diferentes grupos de variáveis que representam múltiplas dimensões do comportamento financeiro.

### 1.6.1 Variáveis Transacionais

Incluem:

* valor da transação (`TransactionAmt`)
* tempo relativo (`TransactionDT`)
* produto ou serviço associado (`ProductCD`)

Essas variáveis representam o comportamento financeiro direto da operação.

### 1.6.2 Informações de Cartão e Endereço

As variáveis:

* `card1` até `card6`
* `addr1` e `addr2`

representam informações relacionadas ao cartão, banco emissor, categoria financeira e localização de cobrança.

### 1.6.7 Features Comportamentais

O dataset contém grupos de variáveis derivadas:

#### C-features

Variáveis de contagem e frequência de entidades relacionadas à transação.

#### D-features

Variáveis temporais relacionadas ao histórico de atividade.

#### M-features

Variáveis de correspondência e validação entre entidades.

#### V-features

Features de engenharia desenvolvidas pela Vesta Corporation, contendo sinais agregados de comportamento, relacionamento e atividade temporal.

Essas variáveis representam o núcleo comportamental do sistema antifraude.


### 1.6.8 Features de Identidade

O dataset de identidade contém:

* tipo de dispositivo
* navegador
* sistema operacional
* resolução de tela
* sinais de autenticação
* informações de fingerprinting

Essas variáveis ajudam a contextualizar quem está realizando a transação e em qual ambiente ela ocorre.


## 1.7 Uso Exclusivo do Conjunto de Treino

Toda a análise exploratória e modelagem inicial serão realizadas exclusivamente utilizando o conjunto de treino (`train`).

O conjunto de teste (`test`) não contém a variável alvo `isFraud`, impossibilitando análises supervisionadas e comparações entre transações fraudulentas e legítimas.

Além disso, utilizar o conjunto de teste durante o processo exploratório poderia introduzir vieses e vazamento de informação (data leakage), comprometendo a capacidade de generalização do modelo.

Por esse motivo, o conjunto de teste será mantido isolado até a etapa final de inferência e avaliação do modelo.

## 1.8 Objetivos Técnicos do Projeto

Os principais objetivos técnicos deste projeto são:

* compreender o comportamento das transações fraudulentas
* identificar padrões temporais, financeiros e comportamentais
* construir features relevantes para detecção de fraude
* desenvolver modelos de classificação supervisionada
* comparar modelos utilizando e não utilizando informações de identidade
* avaliar o impacto das diferentes famílias de features no desempenho final

## Conclusão

A detecção de fraude financeira é um problema complexo, altamente desbalanceado e fortemente dependente de contexto comportamental. O dataset utilizado neste projeto oferece múltiplas camadas de informação — transacional, temporal, geográfica e identitária — permitindo explorar padrões reais de risco presentes em sistemas antifraude modernos.

O foco deste projeto será compreender esses padrões, desenvolver representações relevantes dos dados e construir modelos capazes de identificar transações suspeitas de forma robusta e interpretável.
